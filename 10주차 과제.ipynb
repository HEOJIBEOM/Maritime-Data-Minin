```
import pandas as pd
file = './owid-covid-data.csv'
wanted_df = pd.read_csv(file)
wanted_df.info()
```

```
locations = wanted_df['location']
print(locations)
```

```
locations.unique()
```

```
kor_df = wanted_df[wanted_df.location == 'South Korea']
kor_df.head()

usa_df = wanted_df[wanted_df.location == 'United States']
usa_df.head()
```

```
kor_date_index_df = kor_df.set_index('date')
kor_date_index_df.head()

usa_date_index_df = usa_df.set_index('date')
usa_date_index_df.head()
```

```
vision_df = pd.DataFrame({
    'pig' : [40, 36, 978, 1350, 3420],
    'horse' : [8, 46, 472, 1210, 3830]
}, index = [1990, 1997, 2003, 2009, 2014])

pic_lines = vision_df.plot.line()
```

```
kor_total_cases = kor_date_index_df['total_cases']
kor_total_cases

usa_total_cases = usa_date_index_df['total_cases']
usa_total_cases
```

```
kor_date_index_df.index

usa_date_index_df.index
```

```
target_df = pd.DataFrame({
    'KOR' : kor_total_cases,
    'USA' : usa_total_cases
  }, index = kor_date_index_df.index
)
target_df.head()
```

```
target_df.plot.line()
```

```
target_df['2020-01-01' : '2024-12-31'].plot.line(rot=45)
```

2021년 8월 27일과 2022년 3월 15일 사이 그래프가 급등하는 것을 알 수 있다.

```
kor_population = kor_date_index_df['population']['2020-01-05']
usa_population = usa_date_index_df['population']['2020-01-05']
print('미국 인구(', usa_population, ') / 대한민국 인구(',
      kor_population, ')')
```

```
rate = round((usa_population / kor_population), 2)
rate
```

```
target_wanted_df = pd.DataFrame({
    'KOR' : kor_total_cases * rate,
    'USA' : usa_total_cases
  }, index = kor_date_index_df.index
)

target_wanted_df.iloc[50:56]
target_wanted_df.head()
```

```
target_wanted_df[:].plot.line(rot=45)
```

```
target_wanted_df['2022-01-01':'2022-12-31'].plot.line(rot=45)
```

```
import pandas as pd
path = 'http://covid.ourworldindata.org/data/'
file = path + 'owid-covid-data.csv'
df = pd.read_csv(file)
df.head()
```

```
locations = df['location']
print(locations)
```

```
locations.unique()
```

```
kor_df = df[df.location == 'South Korea']
kor_df.head()

fra_df = df[df.location == 'France']
fra_df.head()
```

```
kor_date_index_df = kor_df.set_index('date')
kor_date_index_df.head()

fra_date_index_df = fra_df.set_index('date')
fra_date_index_df.head()
```

```
kor_total_cases = kor_date_index_df['total_cases']
kor_total_cases

fra_total_cases = fra_date_index_df['total_cases']
fra_total_cases
```

```
kor_date_index_df.index

fra_date_index_df.index
```

```
target_df = pd.DataFrame({
    'KOR' : kor_total_cases,
    'FRA' : fra_total_cases
  }, index = kor_date_index_df.index
)
target_df.head()
```

```
target_df[:].plot.line(rot=45)
```
